# Generative model of Labels, Abilities, and Difficulties (GLAD)

В [семинаре 15](https://github.com/esokolov/ml-course-hse/blob/master/2020-spring/seminars/sem15-em.pdf) мы рассмотрели задачу восстановления истинной разметки по меткам от экспертов (которым мы не можем доверять в полной мере, более того, их предсказания могут расходиться).

Рассмотрим следующую вероятностную модель:

$$ p(L, Z | \alpha, \beta) = \prod_{i=1}^{n} \prod_{j=1}^m \sigma(\alpha_j\beta_i)^{[l_{ij}=z_i]}\sigma(-\alpha_j\beta_i)^{1-[l_{ij}=z_i]} p(z_j)$$

где $l_{ij} -$ ответ $j$-го эксперта на задачу $i$, $z_j -$ истинная разметка, $\alpha_j, \beta_i-$ уровень экспертизы и сложность задачи соответственно. Для более подробного описания модели можно прочитать материалы семинара, а также [оригинальную статью](http://papers.nips.cc/paper/3644-whose-vote-should-count-more-optimal-integration-of-labels-from-labelers-of-unknown-expertise.pdf). Априорное распределение положим равномерным: $p(z_i) = 0.5$.

Обратите внимание, что правдоподобие моделирует не вероятность метки $l_{ij}$ принять значение 1 или 0, а вероятность того, что она равна скрытой переменной $z_i$, т.е. $p(l_{ij} = z_j|z_j, \alpha_j, \beta_i) \neq p(l_{ij} = 1|\alpha_j, \beta_i) $. На M-шаге можете использовать как один шаг градиентного спуска, так и несколько: разумные результаты у вас должны получаться вне зависимости от числа итераций.

$$q(z) = p(z | L, \alpha, \beta) = p(z | \alpha, \beta) \frac{p(L, z | \alpha, \beta)}{\int p(L, z | \alpha, \beta) dz}
       = p(z) \frac{\sigma(\alpha \beta)^{[L=z]}\sigma(-\alpha \beta)^{[L\neq z]}}
       {\sum_{z_0 = 0}^1 p(z_0) \sigma(\alpha \beta)^{[L=z_0]}\sigma(-\alpha \beta)^{[L\neq z_0]}}
       = p(z) \frac{\prod_{i=1} \sigma(\alpha_i \beta)^{[L_i=z]}\sigma(-\alpha_i \beta)^{[L_i\neq z]}}
       {\sum_{z_0 = 0}^1 p(z_0) \left \{ \prod_{i=1}\sigma(\alpha_i \beta)^{[L_i=z_0]}\sigma(-\alpha_i \beta)^{[L_i\neq z_0]} \right \} }
$$

$$
    L = \mathbb{E}_q \log(p(L, z| \alpha, \beta)) = \mathbb{E}_q \log \left( \prod\limits_{j=1}^{n} p(L_j, z_j | \alpha, \beta_j) \right ) = 
    \sum\limits_{k=0}^1 \sum\limits_{j=1}^{n} q(z_j) \log p(L_j, z_j=k | \alpha, \beta_j) = 
    \sum\limits_{k=0}^1 \sum\limits_{j=1}^{n} q(z_j) \log \prod\limits_{i=1}^{m} p(L_{ij}, z_j=k | \alpha_i, \beta_j) =
    \sum\limits_{k=0}^1 \sum\limits_{i=1}^{m} \sum\limits_{j=1}^{n} q(z_j = k) \log p(L_{ij}, z_j=k | \alpha_i, \beta_j)
$$

```python
def softplus(x):
    '''stable version of log(1 + exp(x))'''
    c = (x > 20) * 1.
    return np.log1p(np.exp(x * (1-c)) * (1-c)) + x * c
```

```python
def posterior(alpha, beta, L):
    """ Posterior over true labels z p(z | l, alpha, beta)
    Args:
        alpha: ndarray of shape (n_experts).
        beta: ndarray of shape (n_problems).
        L: ndarray of shape (n_problems, n_experts).
    Returns:
        ndarray of shape (2, n_problems)
        (2 -- for z = 0 and for z = 1, n_problems -- for each problem in data)
    Tip:
        You may use function log_likelihood here
        See page 7 of seminar, the last formula for details
    """
    neg = log_likelihood(alpha, beta, L, np.zeros_like(beta)).sum(axis=1)
    pos = log_likelihood(alpha, beta, L, np.ones_like(beta)).sum(axis=1)

    q_logits = np.vstack([neg, pos])
    return scipy.special.softmax(q_logits, axis=0)


def log_likelihood(alpha, beta, L, z):
    """ p(l=z | z, alpha, beta)
    Args:
        alpha: ndarray of shape (n_experts).
        beta: ndarray of shape (n_problems).
        L: ndarray of shape (n_problems, n_experts).
        z: ndarray of shape (n_problems).
    Returns:
        ndarray of shape (n_problems,)
    Tips:
        See page 7 of seminar, the second formula for details
    """
    indicator = L == z[:, None]
    decart = beta.reshape((-1, 1)) @ alpha.reshape((1, -1))
    return indicator * -softplus(-decart) + ~indicator * -softplus(decart)


def alpha_grad_lb(alpha, beta, L, q):
    """ Gradient of lower bound wrt alpha
    Args:
        alpha: ndarray of shape (n_experts).
        beta: ndarray of shape (n_problems).
        L: ndarray of shape (n_problems, n_experts).
        q: ndarray of shape (2, n_problems).
    Returns:
        ndarray of shape (n_experts,)
    Tips:
        See pages 8-9 of seminar for details
    """
    decart = beta.reshape((-1, 1)) @ alpha.reshape((1, -1))

    indicator = L == np.zeros_like(beta)[:, None]
    logits_neg =  indicator * scipy.special.expit(-decart) - ~indicator * scipy.special.expit(decart)
    logits_pos = ~indicator * scipy.special.expit(-decart) -  indicator * scipy.special.expit(decart)

    logits = np.stack([logits_neg, logits_pos], axis=0)
    return np.sum(q[:, :, None] * beta[None, :, None] * logits, axis=(0, 1))



def logbeta_grad_lb(alpha, beta, L, q):
    """ Gradient of lower bound wrt alpha
    Args:
        alpha: ndarray of shape (n_experts).
        beta: ndarray of shape (n_problems).
        L: ndarray of shape (n_problems, n_experts).
        q: ndarray of shape (2, n_problems).
    Returns:
        ndarray of shape (n_problems,)
    Tips:
        See pages 8-9 of seminar for details
    """
    decart = beta.reshape((-1, 1)) @ alpha.reshape((1, -1))

    indicator = L == np.zeros_like(beta)[:, None]
    logits_neg =  indicator * scipy.special.expit(-decart) - ~indicator * scipy.special.expit(decart)
    logits_pos = ~indicator * scipy.special.expit(-decart) -  indicator * scipy.special.expit(decart)

    logits = np.stack([logits_neg, logits_pos], axis=0)
    return np.sum(q[:, :, None] * alpha[None, None, :] * logits, axis=(0, 2)) * beta
    

def lower_bound(alpha, beta, L, q):
    """ Lower bound
    Args:
        alpha: ndarray of shape (n_experts).
        beta: ndarray of shape (n_problems).
        L: ndarray of shape (n_problems, n_experts).
        q: ndarray of shape (2, n_problems).
    Returns:
        single value, number
    Tips:
        You may use function log_likelihood here
        See page 8 of seminar, the fourth formula for details
    """
    neg = log_likelihood(alpha, beta, L, np.zeros_like(beta))
    pos = log_likelihood(alpha, beta, L, np.ones_like(beta))
    logits = np.stack([neg, pos], axis=0)
    return np.sum(q[:, :, None] * logits)
```

```python
def em(L, n_steps=1000, lr=1e-3, gd_steps=5):
    # initialize parameters
    alpha, logbeta = np.random.randn(m), np.random.randn(n)
    q = np.ones((2, len(logbeta))) * 0.5

    for step in range(n_steps):
        # E-step
        q = posterior(alpha, np.exp(logbeta), L)
        # M-step
        for k in range(gd_steps):
            alpha = lr * alpha_grad_lb(alpha, np.exp(logbeta), L, q)
            logbeta  = lr * logbeta_grad_lb(alpha, np.exp(logbeta), L, q)
     
    return alpha, np.exp(logbeta), q
```

**А что будет, если использовать другой алгоритм для оптимизации M-шага**

```python
def grad(w, L, q):
    alpha = w[:m]
    beta  = np.exp(w[m:])
    alpha_grad = alpha_grad_lb(alpha, np.exp(beta), L, q)
    beta_grad  = logbeta_grad_lb(alpha, np.exp(beta), L, q)

    return np.hstack([alpha_grad, beta_grad])

def em_modified(L, n_steps=100):
    # initialize parameters
    alpha, logbeta = np.random.randn(m), np.random.randn(n)
    weights = np.hstack([alpha, logbeta])
    q = np.ones((2, len(logbeta))) * 0.5

    for step in range(n_steps):
        # E-step
        q = posterior(weights[:m], np.exp(weights[m:]), L)
        # M-step
        res = scipy.optimize.minimize(
            lambda w: lower_bound(w[:m], np.exp(w[m:]), L, q), weights,
            jac=lambda w: grad(w, L, q),
            method='TNC',
        )
        weights = res['x']
    
    return weights[:m], np.exp(weights[m:]), q
```

```python
# (∩ᄑ_ᄑ)⊃━☆ﾟ*･｡*

def determined_em(L, known_inds, known_answ, n_steps=1000, lr=1e-3, gd_steps=5):
    ground_false = known_inds[known_answ == 0]
    ground_truth = known_inds[known_answ == 1]

    # initialize parameters
    alpha, logbeta = np.random.randn(m), np.random.randn(n)
    q = np.ones((2, len(logbeta))) * 0.5

    for step in range(n_steps):
        # E-step
        q = posterior(alpha, np.exp(logbeta), L)
        
        # determination over specific samples
        q[0, ground_false] = 1
        q[1, ground_false] = 0
        q[0, ground_truth] = 0
        q[1, ground_truth] = 1
        
        # M-step
        for k in range(gd_steps):
            alpha = lr * alpha_grad_lb(alpha, np.exp(logbeta), L, q)
            logbeta  = lr * logbeta_grad_lb(alpha, np.exp(logbeta), L, q)
    
    return alpha, np.exp(logbeta), q
```

В качестве бейзлайна для этой задачи мы возьмём способ выравнивания слов по коэффициенту Дайса: слово в исходном языке является переводом слова на целевом языке, если они часто встречаются в одних и тех же предложениях и редко встречаются по отдельности. 

Математически это записывается по аналогии с мерой Жаккара: пусть $c(x,y)$ — число параллельных предложений, в которых есть и $x$ (на исходном языке), и $y$ (на целевом языке), а $c(x)$ и $c(y)$ — суммарное количество предложений, в которых встречается слово $x$ и $y$ соответственно. Тогда $\textrm{Dice}(x,y)=\frac{2 \cdot c(x,y)}{c(x) + c(y)}$ — характеристика «похожести» слов $x$ и $y$. Она равна 1, если слова встречаются только в контексте друг друга (не бывает предложений только со словом $x$ без $y$ в переводе и наоборот), равна 0, если слова никогда не встречаются в параллельных предложениях и находится между пороговыми значениями в остальных случаях.

Чтобы оценить качество модели выравнивания, пользуясь имеющейся разметкой, существует ряд автоматических метрик. Они подразумевают, что в разметке есть два вида выравниваний — «уверенные» (sure) и «возможные» (possible). Обозначим для конкретного предложения первое множество выравниваний $S$, второе — $P$, а предсказанные выравнивания — $A$; причём, в отличие от разметки в файле, $S\subseteq P$. Тогда можно предложить три метрики, используя только операции над этими множествами:

Precision $=\frac{|A\cap P|}{|A|}$. Отражает, какая доля предсказанных нами выравниваний вообще корректна; если мы дадим в качестве ответа все возможные пары слов в предложении, эта метрика сильно просядет.

Recall $=\frac{|A\cap S|}{|S|}$. Эта метрика показывает, какую долю «уверенных» выравниваний мы обнаружили. Если мы попытаемся сделать слишком консервативную модель, которая выдаёт 0 или 1 предсказание на нетривиальных предложениях, полнота получится крайне низкая. 

Alignment Error Rate (AER) $=1-\frac{|A\cap P|+|A\cap S|}{|A|+|S|}$. Метрика является комбинацией двух предыдущих и отслеживает общее качество работы системы, штрафуя оба описанных выше вида нежелаемого поведения модели.

Теперь мы можем перейти к базовой вероятностной модели для выравнивания слов. Пусть $S=(s_1,\ldots,s_n)$ исходное предложение, $T=(t_1,\ldots,t_m)$ — его перевод. В роли латентных переменных будут выступать выравнивания $A=(a_1,\ldots,a_m)$ каждого слова в целевом предложении, причём $a_i\in\{1,\ldots,n\}$ (считаем, что каждое слово в $t$ является переводом какого-то слова из $s$). Параметрами модели является матрица условных вероятностей перевода: каждый её элемент $\theta(y|x)=p(y|x)$ отражает вероятность того, что переводом слова $x$ с исходного языка на целевой является слово $y$ (нормировка, соответственно, совершается по словарю целевого языка). Правдоподобие латентных переменных и предложения на целевом языке в этой модели записывается так:

$$
p(A,T|S)=\prod_{i=1}^m p(a_i)p(t_i|a_i,S)=\prod_{i=1}^m \frac{1}{n}\theta(t_i|s_{a_i}).
$$

*   $\mathcal{M}$ $\rightarrow$ мощность чешского словаря
*   Для прощения записей без индикаторов по трехмерным тензорам будем считать связь $a_i^j(r)$ - бинарным отношением соответствия $i$ токена чешского языка с $j$ токеном английского языка в $r$ предложении.

Считаем, что все распределения над токенами, связями между токенами, априорные распределения - между собой не зависимы в совокупности. Это приводит к тому, что мы имеем право факторизовать распределения совокупностей объектов $p(X)$ как $\prod_{x \in X} p(x)$. Далее не будем на этом останавливаться, лишь подчеркнем, что в любом месте при переходе распределения совокупности случайных величин к произведению вероятностей, мы воспользовались данным предположением.

---

**Рассмотрим постановку самой задачи:** будем рассматривать полную вероятность решения $L = p(T | S)$. Для этого введем переменные $A$ и разобьем оценку полного правдоподобия на Нижнюю баессовскую оценку и дивергенцию Ку́льбака — Ле́йблера:

$$
    log L = \log(p(T | S)) = \left \{ \text{Введем распределение $q$ над случ. элементами $A$} \right \} = \int q(A) \log(p(T | S)) dA =
    \int q(A) \log \left ( \frac{p(T, A | S) q(A)}{p(A | T, S) q(A)} \right ) dA = \\
    = \int q(A) \log \left ( \frac{p(T, A | S)}{q(A)} \right )dA + \int q(A) \log\left ( \frac{q(A)}{p(A | T, S)} \right ) dA = \mathcal{L} + KL(q \| p(A | T, S))
$$

Тогда распишем оба шага:

**E-step:** 
$$
    q(a_i) = p(a_i |t_i, S_{a_i}) = \frac{p(a_i, t_i | S_{a_i})}{p(t_i | S_{a_i})} = \frac{p(a_i) p(t_i | a_i, S_{a_i})}{\sum\limits_{k=1}^n p(a_i^k) p(t_i | a_i^k, S_{a_i^k})} =
    \frac{\theta (t_i | S_{a_i})}{\sum\limits_{k=1}^n \theta (t_i | S_{a_i^k})} = \frac{\langle \textbf{Вероятность соотнесения $t_i$ и $S_{a_i}$}\rangle}
    {\langle \textbf{Сумма вероятностей соотнесения $t_i$ с каждым словом из $S$  }\rangle}
$$

Данная запись валидна при рассмотрении одной единственной пары предложений S и T. В более общем случае мы введем индексацию по корпусу пар текстов $r$. Таким образом на каждом $E$ шаге получаем оценки вероятностей для каждой из связей.

**ELBO-$\mathcal{L}$:** 
$$
\mathcal{L} = \mathbb{E}_q \log \left ( \frac{p(T, A | S)}{q(A)} \right ) = \mathbb{E}_q \log p(T, A | S) - \mathbb{E}_q \log q(A) = \mathbb{E}_q \log \left (\prod\limits_{r=1}^R \prod\limits_{i=1}^{m_r} \frac{1}{n_r}\theta(t_i | S_{a_i(r)}) \right ) - \mathbb{E}_q \log \left (\prod\limits_{r=1}^R \prod\limits_{i=1}^{m_r} \prod\limits_{j=1}^{n_r} q(a_i^j(r))  \right ) = \\ =
\underbrace{\sum\limits_{r=1}^R \sum\limits_{i=1}^{m_r} \mathbb{E}_q \log \theta(t_i | S_{a_i(r)})}_{\text{параметрический лосс от $\theta$}} - \underbrace{\sum\limits_{r=1}^R \sum\limits_{i=1}^{m_r} \sum\limits_{j=1}^{n_r} q(a_i^j(r)) \cdot \log q(a_i^j(r))}_{\text{информация Фишера}} + \underbrace{\sum\limits_{r=1}^R m_r \cdot \log \left(\frac{1}{n_r} \right)}_{\texttt{Const}}
$$

**M-step:** 

$$
    arg\max\limits_\theta \mathcal{L} = arg\max\limits_\theta \, \mathbb{E}_q \log \left ( \frac{p(T, A | S)}{q(A)} \right ) \approxeq arg\max\limits_\theta \, \mathbb{E}_q \log \left ( p(T, A | S) \right ) =
    arg\max\limits_\theta \, \mathbb{E}_q \log \left (\prod\limits_{r=1}^R \prod\limits_{i=1}^{m_r} \frac{1}{n_r}\theta(t_i | S_{a_i(r)}) \right ) \approxeq
    arg\max\limits_\theta \, \mathbb{E}_q \left ( \sum\limits_{r=1}^R \sum\limits_{i=1}^{m_r} \log \theta(t_i | S_{a_i(r)}) \right ) = \\ =
    arg\max\limits_\theta \, \sum\limits_{r=1}^R \sum\limits_{i=1}^{m_r} \mathbb{E}_q \log \theta(t_i | S_{a_i(r)}) = 
    arg\max\limits_\theta \, \sum\limits_{r=1}^R \sum\limits_{i=1}^{m_r} \sum\limits_{j=1}^{n_r} q(a_i^j(r)) \log \theta(t_i | S_{a_i^j(r)})
$$

Тем самым получаем баессовскую нижнюю оценку как сумму по всем предложениям для логарифмов вероятностей каждой из возможных связей (переводов source-target) на вероятность такой связи.

Выведем напрямую градиент для $\theta$:
$\nabla_\theta \mathcal{L} = \sum\limits_{r=1}^R \left \{ \frac{q(a_i^j(r))}{\theta_i^j} \right \}_{i \in \, |T|}^{j \in \, |S|}$

Однако такое определение градиента нам не подходит, так как он не отображает условие $p(t | s) \in [0, 1]$, и обычный градиентый спуск нас не устроит. Более того, если мы применим, например, метод проекции градиента к этой задаче довольно таки скоро му получим неустойчивую ситуацию, где вероятности начнут попеременно обращаться в граничные состояния (0 и 1). Поступим иначе: будем считать, что $\theta_i^j$ моделируют не вероятности, а ее логиты, а сами вероятности приближаются сигмоидой от $\theta_i^j$. Такой принцип кажется не безосновательным, ведь по сути речь идет о бернулевских случайных величинах соответсвия $i \sim j$.

$$\Longrightarrow \nabla_\theta \mathcal{L} = \sum\limits_{r=1}^R \left \{ \frac{q(a_i^j(r))}{\sigma(\theta_i^j)} \sigma^\prime(\theta_i^j) \right \}_{i \in \, |T|}^{j \in \, |S|} = \sum\limits_{l=1}^L \left \{ q(a_i^j(r)) \frac{e^{-\theta_i^j}}{1 + e^{-\theta_i^j}}  \right \}_{i \in \, |T|}^{j \in \, |S|}
$$

Однако теперь получаемые значения не соответсвуют условию нормировки вероятностей к 1, из-за чего нам требуется перевести задачу из разряда градиентного спуска в метод проекции градиента, где проицировать мы будем на множество вещественных покомпонентно неотрицательных матриц $X^{m, n}$, для которых $\forall i \in \{ 1, ..., m\}: \sum\limits_{j = 1}^n X_i^j = 1$

Такой проекцией является матрица $\hat{X}$, отнормированная по строкам

--- 

**А теперь посмотрим на семинар и перестанем тупить:** мы же можем расписать KKT и получить аналитический вывод решения $\rightarrow$

$$\mathbb{L} = \sum\limits_{r=1}^R \sum\limits_{i=1}^{m_r} \sum\limits_{j=1}^{n_r} q(a_i^j(r)) \log \theta(t_i | S_{a_i^j(r)}) + 
\sum\limits_{j=1}^\mathcal{N} \lambda_i \left(1 - \sum_{i=1}^\mathcal{M} \theta_i^j \right) \Longrightarrow 
\frac{\partial}{\partial \theta_i^j}\mathbb{L} = \frac{\sum\limits_{r=1}^R q(a_i^j(r))}{\theta_i^j} \, - \lambda_j = 0 \Longrightarrow
\theta_i^j = \frac{\sum\limits_{l=1}^L q(a_i^j(r))}{\lambda_j}$$

Теперь вернемся к ограничению: $\forall j \in \{1, ..., n\}: \sum\limits_{i=1}^\mathcal{M} \theta_i^j = 1 \Longrightarrow \lambda_j = \sum\limits_{r=1}^R \sum\limits_{i=1}^\mathcal{M} q(a_i^j(r)) \Longrightarrow \theta_i^j = \frac{\sum\limits_{r=1}^R q(a_i^j(r))}{\sum\limits_{r=1}^R \sum\limits_{j=1}^\mathcal{M} q(a_i^j(r))}$

```python
# (•̀ 3 •́)━★☆.*･｡ﾟ
import matplotlib.pyplot as plt
import numpy as np

def make_example_heatmap(ind, sentences, aligner):
    matrix = np.zeros((len(sentences[ind].source_tokens), len(sentences[ind].target_tokens)))
    align = np.array(aligner.align([sentences[ind]])[0])
    matrix[align[:, 0] - 1, align[:, 1] - 1] = 1

    fig, ax = plt.subplots(1, 2)
    t = fig.suptitle('Сравнение alignments и апостериорных распределений')
    t.set_position([.5, .9])
    
    ax[0].imshow(matrix, cmap='hot', interpolation='nearest')
    ax[0].set_title('alignments')
    ax[0].set_xlabel('czech sentence')
    ax[0].set_ylabel('english sentence')

    src_trg_indx = np.ix_(sentences[ind].source_tokens, sentences[ind].target_tokens)
    probas = aligner.translation_probs[src_trg_indx]
    ax[1].imshow(probas, cmap='hot', interpolation='nearest')
    ax[1].set_title('probas')
    ax[1].set_xlabel('czech sentence')
```

**Рассмотрим постановку самой задачи:** будем рассматривать полную вероятность решения $L = p(T | S)$. Для этого введем переменные $A$ и разобьем оценку полного правдоподобия на Нижнюю баессовскую оценку и дивергенцию Ку́льбака — Ле́йблера:

$$
    log L = \log(p(T | S)) = \left \{ \text{Введем распределение $q$ над случ. элементами $A$} \right \} = \int q(A) \log(p(T | S)) dA =
    \int q(A) \log \left ( \frac{p(T, A | S) q(A)}{p(A | T, S) q(A)} \right ) dA = \\
    = \int q(A) \log \left ( \frac{p(T, A | S)}{q(A)} \right )dA + \int q(A) \log\left ( \frac{q(A)}{p(A | T, S)} \right ) dA = \mathcal{L} + KL(q \| p(A | T, S))
$$

Тогда распишем оба шага:

**E-step:** 
$$
    q(a_i) = p(a_i |t_i, S_{a_i}) = \frac{p(a_i, t_i | S_{a_i})}{p(t_i | S_{a_i})} = \frac{p(a_i|m,n)p(t_i| a_i, S)}{\sum\limits_{k=1}^n p(t_i, a_i^k | S_{a_i^k})} = 
    \frac{p(a_i|m,n)p(t_i| a_i, S)}{\sum\limits_{k=1}^n p(a_i^k|m,n) p(t_i | a_i^k, S_{a_i^k})} = 
    \frac{\phi_{m,n}(a_i|i)\theta(t_i|s_{a_i})}{\sum\limits_{k=1}^n \phi_{m,n}(a_i^k|i) \theta(t_i|s_{a_i^k})}
$$

**ELBO-$\mathcal{L}$:** 
$$
\mathcal{L} = \mathbb{E}_q \log \left ( \frac{p(T, A | S)}{q(A)} \right ) = \mathbb{E}_q \log p(T, A | S) - \mathbb{E}_q \log q(A) = \mathbb{E}_q \log \left (\prod\limits_{r=1}^R \prod\limits_{i=1}^{m_r} \frac{1}{n_r}\phi_{m_r}(a_i(r)|i) \cdot \theta(t_i | S_{a_i(r)}) \right ) - \mathbb{E}_q \log \left (\prod\limits_{r=1}^R \prod\limits_{i=1}^{m_r} \prod\limits_{j=1}^{n_r} q(a_i^j(r))  \right ) = \\ =
\underbrace{\sum\limits_{r=1}^R \sum\limits_{i=1}^{m_r} \mathbb{E}_q \log \phi_{m_r, n_r}(a_i^j(r)|i) \cdot \theta(t_i | S_{a_i(r)})}_{\text{параметрический лосс от $\theta$}} - \underbrace{\sum\limits_{r=1}^R \sum\limits_{i=1}^{m_r} \sum\limits_{j=1}^{n_r} q(a_i^j(r)) \cdot \log q(a_i^j(r))}_{\text{информация Фишера}} + \underbrace{\sum\limits_{r=1}^R m_r \cdot \log \left(\frac{1}{n_r} \right)}_{\texttt{Const}}
$$

**M-step:** 

$$
    arg\max\limits_\theta \mathcal{L} = arg\max\limits_\theta \, \mathbb{E}_q \log \left ( \frac{p(T, A | S)}{q(A)} \right ) \approxeq arg\max\limits_\theta \, \mathbb{E}_q \log \left ( p(T, A | S) \right ) =
    arg\max\limits_\theta \, \mathbb{E}_q \log \left (\prod\limits_{r=1}^R \prod\limits_{i=1}^{m_r} \frac{1}{n_r} \phi_{m_r}(a_i(r)|i) \cdot \theta(t_i | S_{a_i(r)}) \right ) \approxeq \\ \approxeq
    arg\max\limits_\theta \, \mathbb{E}_q \left ( \sum\limits_{r=1}^R \sum\limits_{i=1}^{m_r} \log \phi_{m_r}(a_i(r)|i) \cdot \theta(t_i | S_{a_i(r)}) \right ) = 
    arg\max\limits_\theta \, \sum\limits_{r=1}^R \sum\limits_{i=1}^{m_r} \mathbb{E}_q \log \phi_{m_r}(a_i(r)|i) \cdot \theta(t_i | S_{a_i(r)}) = 
    arg\max\limits_\theta \, \sum\limits_{r=1}^R \sum\limits_{i=1}^{m_r} \sum\limits_{j=1}^{n_r} q(a_i^j(r)) \log \phi_{m_r, n_r}(a_i^j(r)|i) \cdot \theta(t_i | S_{a_i^j(r)})
$$

Тем самым получаем баессовскую нижнюю оценку как сумму по всем предложениям для логарифмов вероятностей каждой из возможных связей (переводов source-target) на вероятность такой связи только на этот раз мы учитываем позицию в текстах пары.

**Постановка задачи оптимизации M-шага:**

С помощю $\theta$ и $\phi$ мы моделируем вероятности, а саму оценку мы пытаемся максимизировать. Потому получаем слудующее представление:

$$
\begin{equation*}
\left\{
    \begin{aligned}
    &  \sum\limits_{r=1}^R \sum\limits_{i=1}^{m_r} \sum\limits_{j=1}^{n_r} q(a_i^j(r)) \log \phi_{m_r, n_r}(a_i^j(r)|i) \cdot \theta(t_i | S_{a_i^j(r)}) \rightarrow \max\limits_{\phi, \theta}\\
    & \forall i \in \{1, \cdots, \mathcal{N}\} : \sum_{j=1}^\mathcal{M} \theta_i^j = 1 \\
    & \forall j \in \{1, \cdots, \mathcal{M}\}, \forall m, n : \sum_{i=1}^\mathcal{N} \phi_{m, n}(j | i) = 1
    \end{aligned}
\right.
\end{equation*}
$$

$$\Longrightarrow \mathbb{L} = \sum\limits_{r=1}^R \sum\limits_{i=1}^{m_r} \sum\limits_{j=1}^{n_r} q(a_i^j(r)) \log \phi_{m_r, n_r}(a_i^j(r)|i) +  \sum\limits_{r=1}^R \sum\limits_{i=1}^{m_r} \sum\limits_{j=1}^{n_r} q(a_i^j(r)) \log \theta(t_i | S_{a_i^j(r)}) + \sum_{i=1}^{\mathcal{N}} \lambda_i (1 - \sum_{j=1}^\mathcal{M} \theta_i^j) +
\sum_{(m, n) \in Pairs} \mu_{m, n}(1 - \sum_{j=1}^\mathcal{N} \phi_{m, n}(j | i))$$

Нам удалось факторизовать задачу, получив оптимизации относительно двух независимых групп переменных:

**$\theta$ partial optimization:**

$$
\begin{equation*}
\left\{
    \begin{aligned}
    & \frac{\partial}{\partial \theta_i^j}\mathbb{L}=\frac{\sum\limits_{r=1}^R q(a_i^j(r))}{\theta_i^j} \, - \lambda_j = 0 \\
    & \sum\limits_{i=1}^\mathcal{M} \theta_i^j = 1
    \end{aligned}
\right.
\end{equation*}

\Longrightarrow

\lambda_j = \sum\limits_{r=1}^R \sum\limits_{i=1}^\mathcal{M} q(a_i^j(r)) 

\Longrightarrow

\theta_i^j = \frac{\sum\limits_{r=1}^R q(a_i^j(r))}{\sum\limits_{r=1}^R \sum\limits_{j=1}^\mathcal{M} q(a_i^j(r))}
$$

**$\phi$ partial optimization:**

$$
\begin{equation*}
\left\{
    \begin{aligned}
    & \frac{\partial\mathbb{L}}{\partial \phi_{m, n}(j|i)}=\frac{\sum\limits_{r=1}^R q(a_i^j(r)) \mathbb{I}_{\{m_r = m, \, n_r = n \}}}{\phi_{m, n}(j|i)} \, - \mu_{m,n} = 0 \\
    & \sum\limits_{j=1}^\mathcal{N} \phi_{m, n}(i|j) = 1 \\
    & (m, n) \in Pairs
    \end{aligned}
\right.
\end{equation*}

\Longrightarrow

\mu_{m,n} = \sum\limits_{r=1}^R \sum\limits_{j=1}^\mathcal{N} q(a_i^j(r)) \mathbb{I}_{\{m_r = m, \, n_r = n \}} 

\Longrightarrow

\phi_{m, n}(j|i) = \frac{\sum\limits_{r=1}^R q(a_i^j(r)) \mathbb{I}_{\{m_r = m, \, n_r = n \}}}{\sum\limits_{r=1}^R \sum\limits_{j=1}^\mathcal{N} q(a_i^j(r)) \mathbb{I}_{\{m_r = m, \, n_r = n \}}}
$$
